```python
#First Name is observed to have null values, and are replaced with Mr or Ms depending on gender
df_nan=df_customer[df_customer['First Name'].isnull()==True]
```

```python
df_nan.head(10)
```

```python
def Suffix(row):
    if row['Gender']=='Male':
        return 'Mr'
    else: return 'Ms'
```

```python
df_nan['Suffix']=df_nan.apply(Suffix, axis=1)
```

```python
df_customer['First Name'].fillna(df_nan['Suffix'], inplace=True)
```
