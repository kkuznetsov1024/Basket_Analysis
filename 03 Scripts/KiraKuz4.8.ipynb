```python
#importing libraries
import pandas as pd
import numpy as np
import os
```

```python
#importing the data set for orders and products
ords_prods_merge=pd.read_pickle(r'C:\Users\kirat\Documents\Career Foundy\Python\Instacart Basket Analysis\02 Data\Prepared Data\orders_products_busy.pkl')
```

```python
ords_prods_merge.groupby('department_id').order_number.mean()
```

```python
ords_prods_merge.groupby('department_id').agg({'order_number':['mean','min','max']})
```

When using the entire dataset, we came up with a more accurate result for the mean. We also included all the department. The sample that took only the first rows of the data set did not have every department represented.

```python
ords_prods_merge['max_order']=ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# adding a column for loyalty flag
ords_prods_merge.loc[ords_prods_merge['max_order']>40,'loyalty_flag']='Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order']<=40) & (ords_prods_merge['max_order']>10), 'loyalty_flag']='Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order']<=10, 'loyalty_flag']='New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
ords_prods_merge.groupby('loyalty_flag').prices.mean()
```

On average, it shows that New customers spend the most on products, and Loyal customers spend the least. 

```python
# adding a column for spending flag
ords_prods_merge['mean_spend']=ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spend']<10,'spending_flag']='Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_spend']>=10,'spending_flag']='High spender'
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge['days_since_prior_order'].replace('New Customer', 0, inplace=True)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['days_since_prior_order'].dtypes
```

```python
#adding a column for frequency flag
ords_prods_merge['median_days']=ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days']>20,'frequency_flag']='Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days']>10) & (ords_prods_merge['median_days']<=20),'frequency_flag']='Regular customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days']<=10) & (ords_prods_merge['median_days']>0),'frequency_flag']='Frequent customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days']==0,'frequency_flag']='New customer'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
